# Data Preparation

````python
df.info()
````

````python
df.isnull().sum()
````

````python
missing_percentage = (df.isnull().sum()/df.shape[0])*100
````

````python
missing_percentage[missing_percentage>0].round(2)
````

````python
df.describe().round(2)
````

````python
df.describe(include = ['object'])
````

````python
df.salary_period.unique()
````

````python
df['salary_min'] = np.where(df.salary_period == 'hour', df['salary_min'] * 2080, df['salary_min'])
````

````python
df[df.salary_min == 600]
````

````python
df['salary_raw'] = df['salary_raw'].fillna('Unknown')
df['salary_period'] = df['salary_period'].fillna('Unknown')

df['salary_period_new'] = df['salary_raw'].apply(lambda x: x.split(' ')[-1])
````

````python
df[df.salary_period_new != df.salary_period]
````

````python
df['salary_period_new'].unique()
````

````python
conditions = [
    df['salary_period'] == 'year',
    df['salary_period'] == 'hour', 
    df['salary_period'] == 'day',
    df['salary_period'] == 'month'
]

results = [
    df['salary_min'],            # year -> keep as is
    df['salary_min'],            # hour -> year
    df['salary_min'] * 208,      # day -> year
    df['salary_min'] * 12        # month -> year
]
df['salary_min'] = np.select(conditions, results, default=np.nan)
````

````python
for col in ['salary_max']:
    conditions = [
    df['salary_period'] == 'year',
    df['salary_period'] == 'hour',
    df['salary_period'] == 'day',
    df['salary_period'] == 'month'
    ]
    
    results = [
        df[col],            # year -> keep as is
        df[col] * 2080,     # hour -> year
        df[col] * 208,      # day -> year
        df[col] * 12        # month -> year
    ]
    
    df[col] = np.select(conditions, results, default=np.nan)
````

````python
df.describe()
````

````python
df.fillna(df.mean(numeric_only=True), inplace=True)
````

````python
df.fillna("Unknown", inplace=True)
````

````python
df.isnull().sum()
````

````python
df.duplicated().sum()
````

### Question 6: What is the average salary offered by each company?

````python
df['salary_mid'] = (df['salary_min'] + df['salary_max']) / 2
avg_salary = df.groupby('company')['salary_mid'].mean().sort_values(ascending=True)
avg_salary
````

````python
avg_salary.head(10).plot(kind='barh', figsize=(10,6), color='lightgreen', edgecolor='black')
plt.title("Average salary offered by company")
plt.xlabel("Average salary")
plt.ylabel("Company")
````

### Question 7: Based on the dataset, which skills are most important in hiring?

````python
skills = ['skill_1','skill_2','skill_3','skill_4','skill_5',
          'skill_6','skill_7','skill_8','skill_9','skill_10']

skill_counts = df[skills].stack().loc[lambda x: x != 'Unknown'].value_counts()
skill_counts.head(10)
````

````python
df[skills]
````

````python
skill_counts.head(10).sort_values(ascending=True).plot(kind='barh', figsize=(10,6), color='lightgreen', edgecolor='black')
plt.title("Top 10 most important skills in hiring")
plt.xlabel("Number of hirings")
plt.ylabel("Skill")
````

### Question 8: Which skills are associated with the highest salaries?

````python
skills_salary = pd.concat([
    df[['skill_1', 'salary_mid']].rename(columns={'skill_1': 'skill'}),
    df[['skill_2', 'salary_mid']].rename(columns={'skill_2': 'skill'}),
    df[['skill_3', 'salary_mid']].rename(columns={'skill_3': 'skill'}),
    df[['skill_4', 'salary_mid']].rename(columns={'skill_4': 'skill'}),
    df[['skill_5', 'salary_mid']].rename(columns={'skill_5': 'skill'}),
    df[['skill_6', 'salary_mid']].rename(columns={'skill_6': 'skill'}),
    df[['skill_7', 'salary_mid']].rename(columns={'skill_7': 'skill'}),
    df[['skill_8', 'salary_mid']].rename(columns={'skill_8': 'skill'}),
    df[['skill_9', 'salary_mid']].rename(columns={'skill_9': 'skill'}),
    df[['skill_10', 'salary_mid']].rename(columns={'skill_10': 'skill'})
])

skills_salary = skills_salary[skills_salary['skill'] != 'Unknown']
````

````python
skills_salary
````

````python
avg_salary_by_skill = skills_salary.groupby('skill')['salary_mid'].mean().sort_values(ascending=False)
avg_salary_by_skill
````

````python
avg_salary_by_skill.head(10).plot(kind='bar', figsize=(10,6), color='lightgreen', edgecolor='black')
plt.title("Top 10 skills associated with highest salaries")
plt.xlabel("Average mid salary")
plt.ylabel("Skill")
````

### Question 9: What is the salary difference across states?
